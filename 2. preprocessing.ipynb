```python
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Combining All scraped files

```python
states = os.listdir('media/')

all_df = []

for state in states:
    cities = os.listdir(f'media/{state}')
    state_df_list = []
    for city in cities:
        stations = os.listdir(f'media/{state}/{city}')
        city_df_list = []
        for station in stations:
            station_files = os.listdir(f'media/{state}/{city}/{station}')
            station_df_list = []
            for file in station_files:
                path = f'media/{state}/{city}/{station}/{file}'
                print(path)
                temp_df = pd.read_excel(path)
                df = temp_df[16:][['CENTRAL POLLUTION CONTROL BOARD', 'Unnamed: 2', 'Unnamed: 3']]
                df['state'] = state
                df['city'] = city
                df['station'] = station
                station_df_list.append(df)
            city_df_list.append(pd.concat(station_df_list))
        state_df_list.append(pd.concat(city_df_list))
    all_df.append(pd.concat(state_df_list))

    
df = pd.concat(all_df)
df.reset_index(drop=True, inplace=True)
df.rename(columns={'CENTRAL POLLUTION CONTROL BOARD': 'Date', 'Unnamed: 2':'PM2.5', 'Unnamed: 3': 'PM10'}, inplace=True)
df.replace('None', np.nan, inplace=True)
```
